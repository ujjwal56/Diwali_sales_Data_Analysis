# Gender


```python
sns.countplot(x='Gender',data=df)
```

```python
ax=sns.countplot(x='Gender',data=df)

#to give the labels in male and female
for counts in ax.containers:
    ax.bar_label(counts)
```

```python
# to see how much money spent by male and female
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
money_spent_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender',y='Amount',data=money_spent_gen)
```

```python
money_spent_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
bp=sns.barplot(x='Gender',y='Amount',data=money_spent_gen)
for counts in bp.containers:
    bp.bar_label(counts)
    
```

##### From above plots we can easily say that females are spentmore money on shopping
